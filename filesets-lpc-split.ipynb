```python
import os
import json
```

```python
def split(year,filesets,maxfiles_per_json=1):
    
    nfiles = 0
    names = list(filesets.keys())
    
    for n in names:
        
        files_in_n = len(filesets[n])
        print(n,files_in_n)
        n_jsons = int(files_in_n) 
        
        for i in range(0,n_jsons):
            nfiles += 1
            subfilesets = {}
            subfilesets[n] = [filesets[n][i]]

            this_file = 'infiles-split/'+str(year)+'_'+str(nfiles)+'.json'
            with open(this_file, 'w') as json_file:
                json.dump(subfilesets, json_file)
    print('Created ' + str(nfiles) + ' files')
```

```python
# 2018 
with open('infiles/v2x18_lpc_merged.json') as f:
  filesets = json.load(f)

ccfiles = [n for n in filesets.keys() if "HToCC" in n]
for c in ccfiles:
    filesets.pop(c)

split(2018,filesets)
```
